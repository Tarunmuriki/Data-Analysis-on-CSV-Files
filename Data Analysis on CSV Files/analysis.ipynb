```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load CSV
df = pd.read_csv("data.csv")
print(df.head())

# Basic info
print(df.shape)
print(df.info())

# Group by product and sum sales
sales_summary = df.groupby("Product")["Sales"].sum()
print(sales_summary)

# Plot
sales_summary.plot(kind="bar", figsize=(8,5), title="Sales by Product")
plt.show()
```
